```python
num = df.select_dtypes(exclude='object')
```

```python
num
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
char['brand']=le.fit_transform(char['brand'])
char['processor_brand']=le.fit_transform(char['processor_brand'])
char['processor']=le.fit_transform(char['processor'])
char['gen']=le.fit_transform(char['gen'])
char['os']=le.fit_transform(char['os'])
char['os_bit']=le.fit_transform(char['os_bit'])
char['gpu']=le.fit_transform(char['gpu'])
char['warranty']=le.fit_transform(char['warranty'])
char['Touchscreen']=le.fit_transform(char['Touchscreen'])
char['msoffice']=le.fit_transform(char['msoffice'])
```

```python
char['ram'] = char['ram'].astype(str).str.extract(r'(\d+)').astype('int')

char['ssd'] = char['ssd'].astype(str).str.extract(r'(\d+)').astype('int')

char['hdd'] = char['hdd'].astype(str).str.replace('TB', '000')
char['hdd'] = char['hdd'].str.extract(r'(\d+)').astype('int')

char['rating'] = pd.to_numeric(char['rating'].str.extract(r'(\d+)')[0], errors='coerce')
```

```python
char
```

```python
clean_data = pd.concat([char,num], axis=1)
```

```python
clean_data
```

```python
clean_data.to_csv("laptopPrice_cleaned.csv", index=False)
```
